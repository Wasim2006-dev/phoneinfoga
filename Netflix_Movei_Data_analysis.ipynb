```python
# Ensure 'Vote_Count' is numeric, coercing errors to NaN
df['Vote_Count'] = pd.to_numeric(df['Vote_Count'], errors='coerce')

# Split the 'Genre' column and explode, similar to previous steps for average votes
df_exploded_genres_for_total_votes = df.assign(
    Genre=df['Genre'].astype(str).str.split(',')
).explode('Genre')

# Strip whitespace and drop NaNs
df_exploded_genres_for_total_votes['Genre'] = df_exploded_genres_for_total_votes['Genre'].str.strip()
df_exploded_genres_for_total_votes = df_exploded_genres_for_total_votes.dropna(subset=['Genre', 'Vote_Count'])

# Calculate the total 'Vote_Count' for each genre
genre_votes = df_exploded_genres_for_total_votes.groupby('Genre')['Vote_Count'].sum()

plt.figure(figsize=(12, 6))
genre_votes.sort_values(ascending=False).head(10).plot(kind='bar') # Plot top 10 genres by total votes
plt.title("Top 10 Total Votes by Genre")
plt.xlabel("Genre")
plt.ylabel("Total Vote Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#what movei got the highest popularity,whats its genre

```python
top_movei=df.loc[df['Popularity'].idxmax()]
print("Movei with highest popularity:")
print(f"Title: {top_movei['Title']}")
print(f"Genre: {top_movei['Genre']}")
```

#what movei got the lowest popularity? whats its genre

```python
lowest_movei=df.loc[df['Popularity'].idxmin()]
print("Movie with lowest popularity:")
print("Title:",lowest_movei['Title'])
print("Genre:",lowest_movei['Genre'])
print("Popularity:",lowest_movei['Popularity'])
```

#which year has the most filmmed moveis

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['year'] = df['Release_Date'].dt.year
year_counts = df['year'].value_counts()
print("Year with most movies:", year_counts.idxmax())
print("Number of movies:", year_counts.max())
```
