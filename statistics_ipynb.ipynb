#varience

#using numpy

```python
import numpy as np
data=[2,4,4,5,6,7,8,9]
sample_varience=np.var(data, ddof=1)
print(sample_varience)
pop_varience=np.var(data,ddof=0)
print(pop_varience)
```

#using pandas

```python
import pandas as pd
data=[2,5,6,6,7,8,5,4,9,14,2,10]
Series=pd.Series(data)
print(Series)
sample_vrience=Series.var()
print(sample_vrience)
pop_varience=Series.var(ddof=0)
print(pop_varience)
```

#using scipy

```python
from scipy import stats
data=[7,8,9,1,8,5,2,6,4]
sample_varience=stats.tvar(data)
print(sample_varience)
pop_varience=stats.tvar(data, ddof=0)
print(pop_varience)
```

#standard deviation

#using numpy

```python
import numpy as np
data=[60,65,70,75,80]
sample_std=np.std(data,ddof=1)
print(sample_std)
pop_std=np.std(data,ddof=0)
print(pop_std)
```

#using pandas as pd

```python
import pandas as pd
data=[2,5,6,67,8,5,4,9,14,2,10]
Series=pd.Series(data)
print(Series)
sample_std=Series.std()
print(sample_std)
pop_std=Series.std(ddof=0)
print(pop_std)
```

#using scipy library

```python
from scipy import stats
import math
sample_std=math.sqrt(stats.tvar(data))
print(sample_std)
pop_std=math.sqrt(stats.tvar(data,ddof=0))
print(pop_std)
```
